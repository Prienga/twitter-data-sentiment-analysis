Stemming

Step 5:
**WORD FREQUENCY ANALYSIS**

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
  stemmed_content = re.sub('[^a-zA-Z]',' ',content)
  stemmed_content = stemmed_content.lower()
  stemmed_content = stemmed_content.split()
  stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  stemmed_content = ' '.join(stemmed_content)
  return stemmed_content
```

```python
df['stemmed_content'] = df['text'].apply(stemming)
```

```python
df.head()
```

```python
#separating the data and label
X = df['stemmed_content'].values
Y = df['target'].values
```

```python
print(X)
```

```python
print(Y)
```

```python
df['target'] = df['target'].replace({0: 'negative', 4: 'positive'})
```

```python
df.head()
```

Word cloud visualisation

```python
from wordcloud import WordCloud

positive_text = ' '.join(df[df['target'] == 'positive']['stemmed_content'])
negative_text = ' '.join(df[df['target'] == 'negative']['stemmed_content'])

wordcloud = WordCloud(max_words=100, background_color='white').generate(positive_text)
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Positive Tweets Word Cloud')
plt.show()
```

```python
wordcloud = WordCloud(max_words=100, background_color='white').generate(negative_text)
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Negative Tweets Word Cloud')
plt.show()
```

Step 8:
**SENTIMENT PREDICTION MODEL**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(X_train)
```

```python
print(X_test)
```

Step 9:
**FEATURE EXTRACTION**

```python
#converting text to numerical data
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
#training the ML model
#logistic regression
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train,Y_train)
```

**MODEL EVALUATION**

```python
#accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('accuracy score on the training data : ',training_data_accuracy*100)
```

```python
#accuracy score on testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy score on testing data: ',testing_data_accuracy*100)
```
